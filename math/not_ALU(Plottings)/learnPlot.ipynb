```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x= [1,2,3,4,5]
y=[2,4,6,8,10]
plt.figure(figsize=((5,3)), dpi=100)
plt.plot(x,y, c="red", marker= "s", linewidth= 2, linestyle= "dashed", markeredgecolor= "blue", label= "2x", markersize= 10)
plt.title("My first plot", fontdict= {"fontname":"DejaVu Sans" , "fontsize":20})
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,9,10])
plt.legend()
plt.show()
```

```python
categories= ["a","b","c"]
values =[1,2,3]
bars=plt.bar(categories, values)
bars[0].set_hatch("/")
bars[1].set_hatch("o")
```

```python
gas= pd.read_csv("gas_prices.csv")
```

```python
gas
```

```python
gas["Year"]
gas["Year"][0]
```

```python
columns_plot= ["Australia", "Canada", "France", "Germany"]
for country in gas:
    if country in columns_plot:
        plt.plot(gas.Year, gas[country], label= f"Country: {country}")

plt.legend()
plt.xticks()
plt.title("Evolution of gas prices" , fontdict= {"fontweight": "bold"})
```

Histograms

```python
fifa_data= pd.read_csv("fifa_data.csv")
fifa_data
```

```python
fifa_data.head(10)
```

```python
bins= [40,50,60,70,80,90,100]
plt.hist(fifa_data.Overall, bins= bins , color="green")
plt.xticks(bins)
plt.xlabel("number of players")
plt.ylabel("Overall range")

plt.show()
```

Bar Charts


```python
right= fifa_data.loc[fifa_data["Preferred Foot"]== "Right" ].shape[0]
left= fifa_data.loc[fifa_data["Preferred Foot"]== "Left"].shape[0]
labels= ["left", "right"]
colors=["blue", "green"]
plt.pie([left,right], labels=labels, colors= colors, autopct= "%.2f %%")
plt.title("Percentages of preferred foot of players")
plt.show()
```

scatter charts


```python
study_hours= [3,4,5,6,7,8,9,10]
grades_gpa= [1,2,3,4,4.5,5,2,1]
colors=['red' if x < 3 else 'blue' for x in grades_gpa]
plt.scatter(study_hours, grades_gpa, c=colors)
plt.show()
```
